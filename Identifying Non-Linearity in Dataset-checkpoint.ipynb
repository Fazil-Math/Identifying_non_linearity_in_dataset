```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Here we're creating 4-dimensional spherical input dataset with center (5,5,5,5) and radius 6 units. Spherical shape was specifically chosen as our input data so that the Pearson correlation coefficient between any pair of predictor variable is nearly 0. But centre and radius of the sphere are random choices

```python
x1=[]
x2=[]
x3=[]
x4=[]
def sphere(point):
    if ((point)**2).sum()<=36:
        return True
    return False    
count=0
collect=[]
while count<10000:
    point=5*np.random.randn(4,)
    if sphere(point):
        count+=1
        x1.append(5+point[0])
        x2.append(5+point[1])
        x3.append(5+point[2])
        x4.append(5+point[3])
```

We've created 10000 data points inside the aforementioned higher dimensional sphere. Here we're creating a dataframe with 5 columns, 4 of which are the 4 coordinates of the 10000 data points which we're going to use as predictors and the 5th one is the target variable

We create the target variable by adding the first 4 variables and <b> we're introducing non-linearity by adding square of x1 and also square of x2 to the target variable</b>. Note that <b>x1</b> and <b>x2</b> are the only variables contributing to non-linearity

```python
df=pd.DataFrame()
df['x1']=x1
df['x2']=x2
df['x3']=x3
df['x4']=x4

# up to 1st 4 terms it is linear, but the last 2 terms make the whole thing non-linear
df['tar']=df['x1']+df['x2']+df['x3']+df['x4']+(df['x1']**2)+(df['x2']**2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
line=LinearRegression()
```

```python
X=df.drop('tar',axis=1)
y=df.tar
```

```python
line.fit(X,y)
```

```python
y_pred=line.predict(X)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```

```python
residuals=y-y_pred
```

Linearity of the dataset is questionable if any or all of the following assumptions are violated

- <font color='darkgreen'><b>residuals are normally distributed and centered around 0</b></font>
- <font color='darkgreen'><b>residuals are independent, or in otherwords we won't find any pattern in the error terms</b></font>
- <font color='darkgreen'><b>residuals are homoscedastic, or in otherwords they have uniform variance</b></font>

Observe the following 5 plots
- <font color='navy'><b>residuals vs y_pred</b></font>
- <font color='navy'><b>residuals vs x1</b></font> (contributes to non-linearity)
- <font color='navy'><b>residuals vs x2</b></font> (contributes to non-linearity)
- <font color='navy'><b>residuals vs x3</b></font> (does not contribute to non-linearity)
- <font color='navy'><b>residuals vs x4</b></font> (does not contribute to non-linearity)

Regardless of whether a predictor variable is contributing to non-linearity or not. All the these plots violate all our 3 assumptions about residuals, which implies that <b>residuals vs y_pred</b> is good enough to identify non-linearity in the dataset as a whole, but separate <b>residuals vs predictor</b> plots are not useful in figuring out which predictor variable is contributing to the non-linearity!

```python
plt.figure(dpi=300)
sns.scatterplot(x=y_pred,y=residuals)
sns.lineplot(x=np.linspace(-20,200,100),y=np.zeros(100,),color='Red')
plt.ylabel('residuals')
plt.xlabel('y_pred');
```
